```python
# Importando bibliotecas 
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
```

```python
# Carregando o dataset Diabetes
diabetes = load_diabetes()
X, y = diabetes.data, diabetes.target
```

```python
# Dividindo os dados em treino e teste
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Treinando o modelo Random Forest Regressor
reg = RandomForestRegressor(random_state=42)
reg.fit(X_train_reg, y_train_reg)
```

```python
# Fazendo previsões no conjunto de teste
y_pred_reg = reg.predict(X_test_reg)
```

```python
# Calculando métricas de regressão
mae = mean_absolute_error(y_test_reg, y_pred_reg)
mse = mean_squared_error(y_test_reg, y_pred_reg)
rmse = np.sqrt(mse)
r2 = r2_score(y_test_reg, y_pred_reg)
```

```python
# Exibindo resultados
print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"R² Score: {r2:.4f}")
```

```python
# Gráfico de Dispersão (Regressão)

import matplotlib.pyplot as plt
    
plt.figure(figsize=(10, 6))
plt.scatter(y_test_reg, y_pred_reg, alpha=0.7)
plt.plot([y_test_reg.min(), y_test_reg.max()], [y_test_reg.min(), y_test_reg.max()], 'r--', lw=2)
plt.title("Real vs Previsto")
plt.xlabel("Valores Reais")
plt.ylabel("Valores Previstos")
plt.show()
```
